# CSV

```python
from io import StringIO, BytesIO
#StringIO — Read and write strings as files, (also known as memory files).
```

```python
data = ('col1,col2,col3\n'
               'x,y,1\n'
               'a,b,2\n'
               'c,d,3')
```

```python
type(data)
```

```python
print(data)
```

```python
StringIO()
```

```python
pd.read_csv(StringIO(data))
```

```python
#read from specific columns
df=pd.read_csv(StringIO(data),usecols=['col1','col3'])
```

```python
df.head()
```

```python
#saving the DF in csv file

df.to_csv('Test.csv')
```

```python
df=pd.read_csv(StringIO(data),dtype={'b':int,'c':float,'a':'Int64'})
```

```python
pd.read_csv(StringIO(data), index_col=0) #make the first column as Index
```

# Pickling

All pandas objects are equipped with to_pickle methods which use Python’s cPickle module to save data structures to disk using the pickle format.

```python
df_pickle=df #creating a pickle df
```

```python
df_pickle.to_pickle('df') #reading the pickle df
```
