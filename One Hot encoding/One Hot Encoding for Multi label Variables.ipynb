# one hot encoding

```python
#lets examin how many columns will obtain after one hot encoding these variables
pd.get_dummies(data,drop_first=True).shape
```

```python
#we can see that from just 6 initial categorical variables, we end up with 117 new variables
#What can we do instead?
#solution: We can take the 10 most frequent labels from each variable and the remaimng labels will be skipped
```

```python
#lets find the top 10 most frequent labels/categories for the variable X2
data.X2.value_counts().sort_values(ascending=False).head(15)
```

```python
#lets make a list with the most frequent top 10 categories of the variable and store it in top_10
top_10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
#and now we make the 10 binary variables

for label in top_10:
    data[label]=np.where(data['X2']==label,1,0)
```

```python
#viewing the X2 column with its top_10 labels where binary condition is applied
data[['X2']+top_10].head(20)
```

# get whole set of dummy variables, for all the categorical variables

```python
#defining function with parameters
def one_hot_top_x(df,variable,top_x_labels):
#function to create the dummy variables for the most frequent labels
#we can vary the number of most frequent labels that we encode

    for label in top_x_labels:
        df[variable+'_'+label]=np.where(data[variable]==label,1,0)
        
#read the data again 
data = pd.read_csv(r'C:\Users\Manoj Nahak\Krish Naik Youtube\Feature Engineering\One Hot encoding\mercedes_benz_train.csv', 
                   usecols=['X1','X2','X3','X4','X5','X6'])

#encode X2 into the 10 most frequent categories 
one_hot_top_x(data, 'X2', top_10)
data.head()
```

```python
#find the top_10 categories for X1 variable
top_10=[x for x in data.X1.value_counts().sort_values(ascending=False).head(10).index]

#now create the 10 most frequent dummy variables for X1 using the same function which we created above
one_hot_top_x(data, 'X1', top_10)
data.head()
```
